# EDA

```python
classes_count= df["Revenue"].value_counts()

plt.pie(classes_count,labels=["NO","Yes"],autopct="%1.1f%%")
plt.title("Revenue coming or not?")
```

```python
classes_count= df["Revenue"].value_counts()

ax= sns.barplot(classes_count,color= "blue")
ax.bar_label(ax.containers[0])
```

```python
visitor_count=df["VisitorType"].value_counts()

ax= sns.barplot(visitor_count,color= "red")
ax.bar_label(ax.containers[0])
```

```python
fig, axes = plt.subplots(3,2)

sns.boxplot(
    ax=axes[0,0],
    data=df,
    x="Revenue",
    y="Applicant_Income",
    color="red"
)

sns.boxplot(
    ax=axes[0,1],
    data=df,
    x="Revenue",
    y="Age",
    color="green"
)

sns.boxplot(
    ax=axes[1,0],
    data=df,
    x="Revenue",
    y="Credit_Score",
    color="blue"
)

sns.boxplot(
    ax=axes[1,1],
    data=df,
    x="Revenue",
    y="DTI_Ratio",
    color="gold"
)

sns.boxplot(
    ax=axes[2,0],
    data=df,
    x="Revenue",
    y="Savings",
    color="purple"
)

sns.boxplot(
    ax=axes[2,1],
    data=df,
    x="Revenue",
    y="Loan_Amount",
    color="cyan"
)

plt.tight_layout()
```
